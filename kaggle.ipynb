```python
def holiday(d):
    hol={"2/1/2017","28/1/2017","30/1/2017","31/1/2017","4/4/2017","1/5/2017","30/5/2017","1/7/2017","2/10/2017","5/10/2017","28/10/2017","22/12/2017","25/12/2017","1/1/2018","16/2/2018","19/2/2018","5/4/2018","1/5/2018","18/6/2018","2/7/2018","25/9/2018","1/10/2018","17/10/2018","22/12/2018","25/12/2018"}
    if d in hol:
        return 1
    else:
        return 0
    
def next_day_holiday(d):
    hol={"1/1/2017","27/1/2017","29/1/2017","30/1/2017","3/4/2017","30/4/2017","29/5/2017","30/6/2017","1/10/2017","4/10/2017","27/10/2017","21/12/2017","24/12/2017","31/12/2017","15/2/2018","18/2/2018","4/4/2018","30/4/2018","17/6/2018","1/7/2018","24/9/2018","31/9/2018","16/10/2018","21/12/2018","24/12/2018"}
    if d in hol:
        return 1
    else:
        return 0
    
def norm_time(d):
    if d<=5:
        return 1
    elif d==6:
        return 2
    elif d==7:
        return 3
    elif d<=9:
        return 4
    elif d<=11:
        return 5
    elif d<=16:
        return 6
    elif d<=18:
        return 7
    elif d==19:
        return 8
    elif d==20:
        return 9
    elif d==21:
        return 10
    else:
        return 11
```
