```python
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
def holiday(d):
    hol={"2/1/2017","28/1/2017","30/1/2017","31/1/2017","4/4/2017","1/5/2017","30/5/2017","1/7/2017","2/10/2017","5/10/2017","28/10/2017","22/12/2017","25/12/2017","1/1/2018","16/2/2018","19/2/2018","5/4/2018","1/5/2018","18/6/2018","2/7/2018","25/9/2018","1/10/2018","17/10/2018","22/12/2018","25/12/2018"}
    if d in hol:
        return 1
    else:
        return 0
    
def next_day_holiday(d):
    hol={"1/1/2017","27/1/2017","29/1/2017","30/1/2017","3/4/2017","30/4/2017","29/5/2017","30/6/2017","1/10/2017","4/10/2017","27/10/2017","21/12/2017","24/12/2017","31/12/2017","15/2/2018","18/2/2018","4/4/2018","30/4/2018","17/6/2018","1/7/2018","24/9/2018","31/9/2018","16/10/2018","21/12/2018","24/12/2018"}
    if d in hol:
        return 1
    else:
        return 0
    
def norm_time(d):
    if d<=5:
        return 1
    elif d==6:
        return 2
    elif d==7:
        return 3
    elif d<=9:
        return 4
    elif d<=11:
        return 5
    elif d<=16:
        return 6
    elif d<=18:
        return 7
    elif d==19:
        return 8
    elif d==20:
        return 9
    elif d==21:
        return 10
    else:
        return 11
```

```python
rawdataframe=pd.read_csv("C:\\Users\\fifine\\Desktop\\HKUST\\sem1\\5001\\kaggle\\train.csv",header='infer')
rawdataframe=rawdataframe.set_index("id")
rawdataframe['date1']=rawdataframe["date"].astype(str).str.split( )
rawdataframe['time']=rawdataframe['date1'].apply(lambda x:x[1])
rawdataframe['time']=rawdataframe['time'].apply(lambda x:int(x[:-3]))
rawdataframe['date2']=rawdataframe['date1'].apply(lambda x:x[0])
rawdataframe['weekday']=rawdataframe['date2'].apply(lambda x: datetime.strptime(x, '%d/%m/%Y').isoweekday())
rawdataframe['holiday']=rawdataframe['date2'].apply(holiday)
rawdataframe['next_day_holiday']=rawdataframe['date2'].apply(next_day_holiday)
#a=rawdataframe.groupby(["date2","norm_time"]).mean()
rawdataframe
```

```python
weather=pd.read_csv("C:\\Users\\fifine\\Desktop\\HKUST\\sem1\\5001\\kaggle\\weather.csv",header='infer')
weather.columns=list(range(1,14))

def rain(d):
    if d=="微量":
        return 0
    else:
        return d
    
def norm_rain(d):
    
    if d<=10:
        return 0
    elif d<=30:
        return 1
    elif d<100:
        return 2
    else:
        return 3
    
temp=weather[[1,5,8,10]]
temp[10]=temp[10].apply(rain)
temp[10]=temp[10].astype("float64")
temp
```

```python
temp["norm_rain"]=temp[10].apply(norm_rain)
temp1=rawdataframe.join(temp.set_index(1),on="date2")
temp1.rename(columns={5:'tempreture'})
```

```python
temp1.drop(["date","date1","date2"],axis=1,inplace=True)
```

```python
temp1
```

```python
temp1=temp1.rename(columns={5:'tempreture',8:'wet'})
```

```python
y=temp1.speed
X=temp1.drop(["speed"],axis=1)
```

```python
import xgboost as xgb
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
xgb_model = xgb.XGBRegressor()
kfold = KFold(n_splits=5, shuffle=True, random_state=345) 

param_grid = {
    'learning_rate':[0.30,0.28,0.32],
    'max_depth':[5,6,7],
    'subsample':[0.8,1],  
    'n_estimators': list(range(200, 320, 20))
}
grid_search = GridSearchCV(estimator=xgb_model, scoring='neg_mean_squared_error', param_grid=param_grid, n_jobs=-1, cv=kfold)
grid_result = grid_search.fit(X, y) 
```

```python
print(f'Best: {grid_result.best_score_} using {grid_result.best_params_}','\n')
```

```python
best_estimator=grid_search.best_estimator_
best_estimator
```

```python
xgb_bestmodel=xgb.XGBRegressor(base_score=0.5, booster='gbtree', colsample_bylevel=1,
             colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
             importance_type='gain', interaction_constraints='',
             learning_rate=0.28, max_delta_step=0, max_depth=6,
             min_child_weight=1,monotone_constraints='()',
             n_estimators=200, n_jobs=0, num_parallel_tree=1,
             objective='reg:squarederror', random_state=0, reg_alpha=0,
             reg_lambda=1, scale_pos_weight=1, subsample=1, tree_method='exact',
             validate_parameters=1, verbosity=None)
xgb_bestmodel.fit(X,y)
#result=xgb_bestmodel.predict(testdata)
```
